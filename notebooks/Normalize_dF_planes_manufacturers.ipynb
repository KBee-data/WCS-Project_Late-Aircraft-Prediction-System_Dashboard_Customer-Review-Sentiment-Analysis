## Use regex and replace() to remove unecessary spaces and normalize company names


```python
df_planes_cleaned = df_planes.copy()  # Create a full copy of the original

# remove any double spaces
df_planes_cleaned['manufacturer'] = df_planes_cleaned['manufacturer'].replace(r'\s{2,}', '', regex=True)

# remove common business suffixes (INC, LTD, CORP, CO, INDUSTRIE) at the end of main name, to just keep main name of manufacturer
df_planes_cleaned['manufacturer'] = df_planes_cleaned['manufacturer'].replace(r'(INC|LTD|CORP|CO|INDUSTRIE)\s*$', '', regex=True)

# remove any single character (\w$) if it's at the end of the leftover string with manufacturer name
df_planes_cleaned['manufacturer'] = df_planes_cleaned['manufacturer'].replace(r'\s\w$', '', regex=True)

# remove any leftover trailing spaces
df_planes_cleaned['manufacturer'] = df_planes_cleaned['manufacturer'].replace(r'\s$', '', regex=True)
```

```python
df_planes_cleaned.to_csv('df_planes_cleaned.csv', index=False)
```
